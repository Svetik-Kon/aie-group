```python
# 1. Импорт библиотеки NumPy с общепринятым сокращением.
import numpy as np
```

```python
# 2. Примеры создания массивов:

# Из обычного списка
python_list = [1, 2, 3, 4, 5]
array_from_list = np.array(python_list)
print("Массив из списка:", array_from_list)

# Использование zeros - массив нулей
zeros_array = np.zeros((2, 3))
print("Массив нулей 2x3:")
print(zeros_array)

# Использование ones - массив единиц
ones_array = np.ones((3, 2))
print("Массив единиц 3x2:")
print(ones_array)

# Использование arange - последовательность чисел
range_array = np.arange(0, 10, 2)  # от 0 до 10 с шагом 2
print("Массив от 0 до 10 с шагом 2:", range_array)
```

```python
# 3. Демонстрация атрибутов:
print("Демонстрационный массив:")
print(zeros_array)

# Вывод формы массива (shape)
print("Форма массива (shape):", zeros_array.shape)

# Вывод типа данных (dtype)
print("Тип данных (dtype):", zeros_array.dtype)
```

```python
# 4. Примеры индексирования и срезов:
# Создаем массив для работы с индексами
arr = np.array([[10, 11, 12, 13],
                [20, 21, 22, 23],
                [30, 31, 32, 33]])

print("Исходный массив:")
print(arr)

# Получение отдельного элемента
element = arr[1, 2]  # строка 1, столбец 2
print("Элемент [1, 2]:", element)

# Извлечение подмассива с помощью среза
sub_array = arr[0:2, 1:3]  # строки 0-1, столбцы 1-2
print("Подмассив [0:2, 1:3]:")
print(sub_array)

# Другие примеры срезов
first_row = arr[0, :]  # первая строка, все столбцы
print("Первая строка:", first_row)

every_second_row = arr[::2, :]  # каждая вторая строка
print("Каждая вторая строка:")
print(every_second_row)
```

```python
# 5. Примеры векторных операций и базовых статистик:
# Создаем массивы для операций
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 4, 6, 8, 10])

print("Массив A:", a)
print("Массив B:", b)

# Поэлементное сложение
addition = a + b
print("Поэлементное сложение A + B:", addition)

# Поэлементное умножение
multiplication = a * b
print("Поэлементное умножение A * B:", multiplication)

# Вычисление статистик для массива A
print("\nСтатистики для массива A:")
print("Среднее значение (mean):", np.mean(a))
print("Сумма элементов (sum):", np.sum(a))
print("Минимальное значение (min):", np.min(a))
```
